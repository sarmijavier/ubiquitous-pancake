```python
# 2. Define your "Index" column (the one you want to keep but not sample)
id_col_name = names(df)[1] # Takes the first column name as a String

# 3. Get all OTHER column names to sample from
all_data_cols = names(df)[2:end]

percentages = [0.30, 0.60, 0.90]
num_iterations = 100

for p in percentages
    for i in 1:num_iterations
        num_to_select = Int(round(length(all_data_cols) * p))
        
        # Sample the column names
        selected_data_cols = sample(all_data_cols, num_to_select, replace=false)
        
        # Combine the ID column name with the sampled names
        # We use [id_col_name; selected_data_cols] or vcat()
        final_selection = vcat(id_col_name, selected_data_cols)
        
        # Create the subset
        sub_df = df[:, final_selection]
        
        # 4. Save with a unique name
        # Example: sample_30_run_1.tsv
        filename = "sample_$(Int(p*100))_run_$i.tsv"
        # CSV.write(filename, sub_df, delim='\t') # Uncomment to save
        CSV.write("../random_columns/$(Int(p*100))/$(filename)", sub_df, delim='\t')
        println("Saved: $filename")
        
        if i == 1 # Just to show progress for the first run of each %
            println("Created $(p*100)% subset (Run $i) with $(ncol(sub_df)) columns.")
        end
    end
end
```

```python

percentages = [0.30, 0.60, 0.90]
num_iterations = 100

for p in percentages
    for i in 1:num_iterations
		filename = "sample_$(Int(p*100))_run_$i.tsv"
		df = CSV.read("../random_columns/$(Int(p*100))/$(filename)", DataFrame, delim='\t')	
		combined_df = leftjoin(df, meta_sub, on = :Sample)

		# Create the Control group (where is_healthy is 1)
		df_healthy = filter(row -> row.is_healthy == 1, combined_df)

		# Create the Case/Patient group (where is_healthy is 0)
		df_unhealthy = filter(row -> row.is_healthy == 0, combined_df)

		# Remove the column from the control group
		select!(df_healthy, Not(:is_healthy))

		# Remove the column from the patient group
		select!(df_unhealthy, Not(:is_healthy))
		
		healthy = FlashWeave.learn_network(
			Matrix(df_healthy[:, Not(:Sample)]),
			sensitive=true,
			heterogeneous=false
		)	

		unhealthy = FlashWeave.learn_network(
			Matrix(df_unhealthy[:, Not(:Sample)]),
			sensitive=true,
			heterogeneous=false
		)

		filename_network_healthy = "sample_$(Int(p*100))_run_network_healthy_$i.edgelist"
		save_network("../random_columns/networks/$(Int(p*100))/$(filename_network_healthy)", healthy)
		filename_network_unhealthy = "sample_$(Int(p*100))_run_network_unhealthy_$i.edgelist"
		save_network("../random_columns/networks/$(Int(p*100))/$(filename_network_unhealthy)", unhealthy)
    end
end
```

```python
import pandas as pd
import os

percentages = [30, 60, 90]
num_iterations = 100

results = []

for p in percentages:
    print(f"Processing {p}% subsets...")
    
    for i in range(1, num_iterations + 1):
        filename = f"sample_{p}_run_network_unhealthy_{i}.edgelist"
        path = f"../random_columns/networks/{p}/{filename}"
        
        if not os.path.isfile(path):
            continue
        
        print(f"Reading: {filename}")
        
        try:
            # Read raw (no assumptions about columns)
            df = pd.read_csv(path , sep='\t', names=['source', 'target', 'weight'])
            # Extract weight column safely
            df['weight'] = df['weight'].astype(float)
            
            sum_pos = df.loc[df['weight'] > 0, 'weight'].sum()

            # sum_neg = Σ ω- (sum of absolute values of negative weights)
            sum_neg = df.loc[df['weight'] < 0, 'weight'].abs().sum()

            # 3. Calculate Inferred Net Interaction
            rho_inferred = (sum_pos - sum_neg) / (sum_pos + sum_neg)
                        
            results.append((p, i, rho_inferred))
        
        except Exception as e:
            print(f"Error reading {filename}: {e}")

all_results = pd.DataFrame(results, columns=["Percentage", "Iteration", "Rho"])
all_results.to_csv('results/df_unhealthy_30_60_90_WANG_ESRD_2020_columns.tsv', sep='\t', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os

# 1. Load and Preprocess Data
all_data = []

# Update this path to where your files are stored
file_list = glob.glob("results/*.tsv")

for file_path in file_list:
    filename = os.path.basename(file_path)
    parts = filename.split('_')
    
    # Extract metadata from filename
    # Example: df_healthy_30_60_90_YACHIDA_CRC_2019_rows.tsv
    status = parts[1]      # healthy / unhealthy
    dimension = parts[-1].replace('.tsv', '') # rows / columns
    disease = parts[5]     # e.g., YACHIDA, iHMP, WANG
    
    df = pd.read_csv(file_path, sep='\t')
    df['Status'] = status.capitalize()
    df['Disease'] = disease
    df['Dimension'] = dimension
    all_data.append(df)

master_df = pd.concat(all_data, ignore_index=True)

# 2. Plotting Function
def plot_dimension(df, dim_name):
    subset = df[df['Dimension'] == dim_name]
    diseases = subset['Disease'].unique()
    
    fig, axes = plt.subplots(len(diseases), 1, figsize=(10, 3 * len(diseases)), sharex=True)
    
    # If only one disease, axes isn't a list
    if len(diseases) == 1: axes = [axes]
    
    for i, disease in enumerate(diseases):
        data_plot = subset[subset['Disease'] == disease]
        
        sns.boxplot(
            data=data_plot, 
            x='Percentage', 
            y='Rho', 
            hue='Status',
            ax=axes[i],
            palette={'Healthy': '#1f77b4', 'Unhealthy': '#d62728'},
            width=0.6
        )
        
        axes[i].set_title(f"Disease Study: {disease}", loc='left', fontweight='bold')
        axes[i].set_ylabel("Inferred Net Interaction")
        axes[i].legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')
        
    plt.xlabel("Subset Fraction (%)")
    plt.tight_layout()
    plt.savefig(f"plot_{dim_name}.png", dpi=300)
    plt.show()

# 3. Execute
plot_dimension(master_df, 'rows')
plot_dimension(master_df, 'columns')
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import glob
import os
import numpy as np
```

```python
def plot_degree_distributions(file_pattern):
    files = sorted(glob.glob(file_pattern))[:100] # Ensure we only take the first 100

    # 2. Initialize the grid (10 rows, 10 columns)
    fig, axes = plt.subplots(10, 10, figsize=(20, 20))
    axes = axes.flatten() # Flatten to 1D array for easy iteration
    edge_data_schema = [('weight', float)]
    for i, file_path in enumerate(files):
        # Load the edgelist
        # Note: adjust create_using if your graph is directed
        #print(f"Processing: {file_path}")

        G = nx.read_edgelist(file_path, data=edge_data_schema)
        
        # Get the degree of all nodes
        degrees = [d for n, d in G.degree()]
        
        # Plot on the specific subplot
        ax = axes[i]
        ax.hist(degrees, bins=15, color='skyblue', edgecolor='black')
        
        # Clean up formatting for readability
        file_num = os.path.basename(file_path).split('_')[-1].replace('.edgelist', '')
        ax.set_title(f"Run {file_num}", fontsize=8)
        ax.tick_params(labelsize=6)

    # 3. Final layout adjustments
    plt.tight_layout()
    plt.show()
```

```python
def plot_to_ax(G, ax, file_num, num_bins=10):
    degree_sequence = [d for n, d in G.degree() if d > 0]
    if not degree_sequence: return # Skip empty graphs
    
    num_nodes = G.number_of_nodes()
    min_deg, max_deg = min(degree_sequence), max(degree_sequence)
    
    # Handle cases where all nodes have the same degree
    if min_deg == max_deg:
        ax.hist(degree_sequence, bins=1)
        ax.set_title(f"Run {file_num}: Constant", fontsize=8)
        return

    bins = np.logspace(np.log10(min_deg), np.log10(max_deg), num_bins + 1)
    counts, bin_edges = np.histogram(degree_sequence, bins=bins)
    
    # Normalize
    bin_widths = np.diff(bin_edges)
    density = counts / (num_nodes * bin_widths)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Linear regression on log-log data
    idx = density > 0
    if np.any(idx):
        log_k = np.log10(bin_centers[idx])
        log_P = np.log10(density[idx])
        
        # We need at least 2 points to fit a line
        if len(log_k) > 1:
            m, b = np.polyfit(log_k, log_P, 1)
            theoretical_P = 10**(m * np.log10(bin_centers) + b)
            ax.plot(bin_centers, theoretical_P, 'k-', alpha=0.7, label=f'$\\gamma={-m:.1f}$')
        
        ax.scatter(bin_centers[idx], density[idx], s=10, alpha=0.6)

    # Styling the subplot
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_title(f"Run {file_num}", fontsize=8)
    ax.tick_params(labelsize=6)
    # Only show legend if we managed to fit a line
    if len(ax.get_lines()) > 0:
        ax.legend(fontsize=5, loc='upper right')

# --- Main Execution ---
def plot_degree_distributions(file_pattern):

    files = sorted(glob.glob(file_pattern))[:100]

    fig, axes = plt.subplots(10, 10, figsize=(25, 25))
    axes = axes.flatten()

    # Note: Using the weight fix from our previous step
    edge_schema = [('weight', float)]

    for i, file_path in enumerate(files):
        G = nx.read_edgelist(file_path, data=edge_schema)
        file_id = os.path.basename(file_path).split('_')[-1].split('.')[0]
        
        plot_to_ax(G, axes[i], file_id)

    # Global labels to save space
    fig.text(0.5, 0.01, 'Degree (k)', ha='center', fontsize=20)
    fig.text(0.01, 0.5, 'P(k)', va='center', rotation='vertical', fontsize=20)

    plt.tight_layout(rect=[0.02, 0.02, 1, 1])
    plt.show()
```
