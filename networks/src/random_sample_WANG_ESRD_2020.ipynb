```python
import FlashWeave
using CSV
using DataFrames
using StatsBase, Random
```

```python
df = CSV.read("../../processed_data/WANG_ESRD_2020/genera.tsv", DataFrame, delim='\t')
```

```python
total_rows = nrow(df)

percentages = [0.30, 0.60, 0.90]
num_iterations = 100

for p in percentages
    # Create the directory if it doesn't exist (optional but helpful)
    mkpath("../YACHIDA_CRC_2019/random_rows/$(Int(p*100))")
    
    for i in 1:num_iterations
        # Calculate how many rows to pick
        num_to_select = Int(round(total_rows * p))
        
        # 2. Sample random ROW indices
        selected_row_indices = sample(1:total_rows, num_to_select, replace=false)
        
        # 3. Create the subset (All columns, but only selected rows)
        # The syntax df[rows, columns] uses : for all columns
        sub_df = df[selected_row_indices, :]
        
        # 4. Save with a unique name
        filename = "sample_$(Int(p*100))_run_$i.tsv"
        
        # Save to the specific row-based directory
        path = "../random_rows/$(Int(p*100))/$(filename)"
        CSV.write(path, sub_df, delim='\t')
        
        if i == 1
            println("Created $(p*100)% Row Subset (Run $i) with $(nrow(sub_df)) samples and $(ncol(sub_df)) columns.")
        end
    end
end
```

```python
metadata_small = select(df_main_meta, :Sample, :"Study.Group")
# Create a column for each unique value in Study.Group
meta_sub = select(df_main_meta, :Sample, :"Study.Group")
# One-Hot Encode "Control" (1 if Control, 0 if not)
meta_sub.is_healthy = [g == "Control" ? 1 : 0 for g in meta_sub[: , :"Study.Group"]]

meta_sub = select(meta_sub, :Sample, :is_healthy)
```

```python

percentages = [0.60, 0.90]
num_iterations = 100

for p in percentages
    for i in 1:num_iterations
		filename = "sample_$(Int(p*100))_run_$i.tsv"
		df = CSV.read("../random_rows/$(Int(p*100))/$(filename)", DataFrame, delim='\t')	
		combined_df = leftjoin(df, meta_sub, on = :Sample)

		# Create the Control group (where is_healthy is 1)
		df_healthy = filter(row -> row.is_healthy == 1, combined_df)

		# Create the Case/Patient group (where is_healthy is 0)
		df_unhealthy = filter(row -> row.is_healthy == 0, combined_df)

		# Remove the column from the control group
		select!(df_healthy, Not(:is_healthy))

		# Remove the column from the patient group
		select!(df_unhealthy, Not(:is_healthy))
		
		healthy = FlashWeave.learn_network(
			Matrix(df_healthy[:, Not(:Sample)]),
			sensitive=true,
			heterogeneous=false
		)	

		unhealthy = FlashWeave.learn_network(
			Matrix(df_unhealthy[:, Not(:Sample)]),
			sensitive=true,
			heterogeneous=false
		)

		filename_network_healthy = "sample_$(Int(p*100))_run_network_healthy_$i.edgelist"
		save_network("../random_rows/networks/$(Int(p*100))/$(filename_network_healthy)", healthy)
		filename_network_unhealthy = "sample_$(Int(p*100))_run_network_unhealthy_$i.edgelist"
		save_network("../random_rows/networks/$(Int(p*100))/$(filename_network_unhealthy)", unhealthy)
    end
end
	

```

```python
import pandas as pd
import os

percentages = [60, 90]
num_iterations = 100

results = []

for p in percentages:
    print(f"Processing {p}% subsets...")
    
    for i in range(1, num_iterations + 1):
        filename = f"sample_{p}_run_network_unhealthy_{i}.edgelist"
        path = f"../WANG_ESRD_2020/random_rows/networks/{p}/{filename}"
        
        if not os.path.isfile(path):
            continue
        
        print(f"Reading: {filename}")
        
        try:
            # Read raw (no assumptions about columns)
            df = pd.read_csv(path , sep='\t', names=['source', 'target', 'weight'])
            # Extract weight column safely
            df['weight'] = df['weight'].astype(float)
            
            sum_pos = df.loc[df['weight'] > 0, 'weight'].sum()

            # sum_neg = Σ ω- (sum of absolute values of negative weights)
            sum_neg = df.loc[df['weight'] < 0, 'weight'].abs().sum()

            # 3. Calculate Inferred Net Interaction
            rho_inferred = (sum_pos - sum_neg) / (sum_pos + sum_neg)
                        
            results.append((p, i, rho_inferred))
        
        except Exception as e:
            print(f"Error reading {filename}: {e}")

all_results = pd.DataFrame(results, columns=["Percentage", "Iteration", "Rho"])
all_results.to_csv('results/df_unhealthy_60_90_WANG_ESRD_2020_rows.tsv', sep='\t', index=False)
```
