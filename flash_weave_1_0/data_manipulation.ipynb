```python
def create_random_subsets_rows(df, dataset_name):
    total_rows = df.shape[0]

    percentages = [0.30, 0.60, 0.90]
    num_iterations = 100

    for p in percentages:
        # Create directory
        dir_path_healthy = f"{dataset_name}/random_rows/{int(p*100)}/healthy"
        dir_path_unhealthy = f"{dataset_name}/random_rows/{int(p*100)}/unhealthy"

        df_main_meta = pd.read_csv(f"../processed_data/{dataset_name}/metadata.tsv", sep='\t')
        os.makedirs(dir_path_healthy, exist_ok=True)
        os.makedirs(dir_path_unhealthy, exist_ok=True)

        if dataset_name == "iHMP_IBDMDB_2019":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()
            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'nonIBD').astype(int)
            # 3. Finalize the meta_sub selection
            
        elif dataset_name == "WANG_ESRD_2020":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Control').astype(int)
        
        elif dataset_name == "YACHIDA_CRC_2019":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Healthy').astype(int)

        meta_sub = meta_sub[['Sample', 'is_healthy']]

        for i in range(1, num_iterations + 1):
            # 1. Number of rows to select
            num_to_select = int(round(total_rows * p))

            # 2. Sample random row indices (no replacement)
            selected_indices = np.random.choice(
                df.index,
                size=num_to_select,
                replace=False
            )

            # 3. Subset dataframe
            sub_df = df.loc[selected_indices, :]

            combined_df = pd.merge(sub_df, meta_sub, on='Sample', how='left')

            # get healthy and unhealthy subsets
            df_healthy = combined_df[combined_df['is_healthy'] == 1].copy()
            df_unhealthy = combined_df[combined_df['is_healthy'] == 0].copy()

            # Drop the helper column from both
            df_healthy = df_healthy.drop(columns=['is_healthy'])
            df_unhealthy = df_unhealthy.drop(columns=['is_healthy'])

            #4. Save file
            filename_healthy = f"sample_{int(p*100)}_run_{i}.tsv"
            filename_unhealthy = f"sample_{int(p*100)}_run_{i}.tsv"
            path_healthy = os.path.join(dir_path_healthy, filename_healthy)
            path_unhealthy = os.path.join(dir_path_unhealthy, filename_unhealthy)
            
            df_healthy.set_index('Sample', inplace=True)
            df_unhealthy.set_index('Sample', inplace=True)


            df_healthy.to_csv(path_healthy, sep="\t", index=True)
            df_unhealthy.to_csv(path_unhealthy, sep="\t", index=True)

            # Print only first iteration
            if i == 1:
                print(
                    f"{dataset_name}: "
                    f"Created {int(p*100)}% Row Subset (Run {i}) "
                    f"with {sub_df.shape[0]} samples and {sub_df.shape[1]} columns."
                )
```

```python
create_random_subsets_rows(df_iHMP, "iHMP_IBDMDB_2019")
create_random_subsets_rows(df_WANG, "WANG_ESRD_2020")
create_random_subsets_rows(df_YACHIDA, "YACHIDA_CRC_2019")
```

```python
# cols In this case means samples. 
def create_random_subsets_cols(df, dataset_name):
	# 1. ID column (first column)
	id_col_name = df.columns[0]

	# 2. All other columns
	all_data_cols = df.columns[1:]

	percentages = [0.30, 0.60, 0.90]
	num_iterations = 100

	for p in percentages:
		# Create directory
		dir_path_healthy = f"{dataset_name}/random_columns/{int(p*100)}/healthy"
		dir_path_unhealthy = f"{dataset_name}/random_columns/{int(p*100)}/unhealthy"

		df_main_meta = pd.read_csv(f"../processed_data/{dataset_name}/metadata.tsv", sep='\t')
		os.makedirs(dir_path_healthy, exist_ok=True)
		os.makedirs(dir_path_unhealthy, exist_ok=True)

		if dataset_name == "iHMP_IBDMDB_2019":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()
			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'nonIBD').astype(int)
			# 3. Finalize the meta_sub selection
			
		elif dataset_name == "WANG_ESRD_2020":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Control').astype(int)

		elif dataset_name == "YACHIDA_CRC_2019":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Healthy').astype(int)

		meta_sub = meta_sub[['Sample', 'is_healthy']]

		for i in range(1, num_iterations + 1):
			# 3. Number of columns to select
			num_to_select = int(round(len(all_data_cols) * p))

			# 4. Sample column names (no replacement)
			selected_data_cols = np.random.choice(
				all_data_cols,
				size=num_to_select,
				replace=False
			)

			# 5. Combine ID column + sampled columns
			final_selection = [id_col_name] + list(selected_data_cols)

			# 6. Subset dataframe
			sub_df = df.loc[:, final_selection]

			combined_df = pd.merge(sub_df, meta_sub, on='Sample', how='left')

			# get healthy and unhealthy subsets
			df_healthy = combined_df[combined_df['is_healthy'] == 1].copy()
			df_unhealthy = combined_df[combined_df['is_healthy'] == 0].copy()

			# Drop the helper column from both
			df_healthy = df_healthy.drop(columns=['is_healthy'])
			df_unhealthy = df_unhealthy.drop(columns=['is_healthy'])

			# 4. Save file
			filename_healthy = f"sample_{int(p*100)}_run_{i}.tsv"
			filename_unhealthy = f"sample_{int(p*100)}_run_{i}.tsv"
			path_healthy = os.path.join(dir_path_healthy, filename_healthy)
			path_unhealthy = os.path.join(dir_path_unhealthy, filename_unhealthy)

			df_healthy.set_index('Sample', inplace=True)
			df_unhealthy.set_index('Sample', inplace=True)
			df_healthy.to_csv(path_healthy, sep="\t", index=True)
			df_unhealthy.to_csv(path_unhealthy, sep="\t", index=True)

			# Print progress
			if i == 1:
				print(
					f"Created {int(p*100)}% subset (Run {i}) "
					f"with {sub_df.shape[1]} columns."
				)
```

```python
create_random_subsets_cols(df_iHMP, "iHMP_IBDMDB_2019")
create_random_subsets_cols(df_WANG, "WANG_ESRD_2020")
create_random_subsets_cols(df_YACHIDA, "YACHIDA_CRC_2019")
```

```python
def validate_random_rows_samples(base_dir, metadata_paths, verbose=True):
	results = []
	for dataset_name, meta_path in metadata_paths.items():
		if not os.path.exists(meta_path):
			print(f"Metadata not found for {dataset_name}: {meta_path}")
			continue

		meta_samples = pd.read_csv(meta_path, sep='\t', usecols=['Sample'])['Sample'].astype(str)
		meta_sample_set = set(meta_samples)

		dataset_random_rows = os.path.join(base_dir, dataset_name, "random_rows")
		if not os.path.isdir(dataset_random_rows):
			print(f"Missing random_rows folder for {dataset_name}: {dataset_random_rows}")
			continue

		for pct in ["30", "60", "90"]:
			for health_status in ["healthy", "unhealthy"]:
				folder = os.path.join(dataset_random_rows, pct, health_status)
				if not os.path.isdir(folder):
					print(f"Missing folder: {folder}")
					continue

				for fname in sorted(os.listdir(folder)):
					if not fname.endswith(".tsv"):
						continue

					path = os.path.join(folder, fname)
					df = pd.read_csv(path, sep='\t', dtype=str, low_memory=False)



					if "Sample" in df.columns:
						#"Study.Group"
						sample_ids = df["Sample"].astype(str)
					else:
						sample_ids = df.iloc[:, 0].astype(str)

					unique_samples = sample_ids.unique()
					missing_samples = [s for s in unique_samples if s not in meta_sample_set]

					results.append({
						"dataset": dataset_name,
						"percentage": pct,
						"group": health_status,
						"file": fname,
						"n_samples": len(unique_samples),
						"n_missing": len(missing_samples),
						"missing_samples": missing_samples,
					})

					if verbose and missing_samples:
						print(
							f"{dataset_name} {pct}% {health_status} {fname}: "
							f"{len(missing_samples)} missing sample IDs"
						)

	return pd.DataFrame(results)

metadata_paths = {
	"WANG_ESRD_2020": "../processed_data/WANG_ESRD_2020/metadata.tsv",
	"iHMP_IBDMDB_2019": "../processed_data/iHMP_IBDMDB_2019/metadata.tsv",
	"YACHIDA_CRC_2019": "../processed_data/YACHIDA_CRC_2019/metadata.tsv",
}

results_df = validate_random_rows_samples("./", metadata_paths)
results_df.head()
```

```python
def _get_health_map(meta_df, dataset_name):
	meta_df = meta_df[['Sample', 'Study.Group']].copy()
	if dataset_name == "iHMP_IBDMDB_2019":
		meta_df['is_healthy'] = (meta_df['Study.Group'] == 'nonIBD').astype(int)
	elif dataset_name == "WANG_ESRD_2020":
		meta_df['is_healthy'] = (meta_df['Study.Group'] == 'Control').astype(int)
	elif dataset_name == "YACHIDA_CRC_2019":
		meta_df['is_healthy'] = (meta_df['Study.Group'] == 'Healthy').astype(int)
	else:
		raise ValueError(f"Unknown dataset: {dataset_name}")
	return meta_df.set_index('Sample')['is_healthy'].to_dict()

def validate_random_rows_samples(base_dir, metadata_paths, verbose=True):
	results = []
	for dataset_name, meta_path in metadata_paths.items():
		if not os.path.exists(meta_path):
			print(f"Metadata not found for {dataset_name}: {meta_path}")
			continue

		meta_df = pd.read_csv(meta_path, sep='\t', usecols=['Sample', 'Study.Group'], dtype=str)
		meta_df['Sample'] = meta_df['Sample'].astype(str)
		meta_sample_set = set(meta_df['Sample'])
		health_map = _get_health_map(meta_df, dataset_name)

		dataset_random_rows = os.path.join(base_dir, dataset_name, "random_rows")
		if not os.path.isdir(dataset_random_rows):
			print(f"Missing random_rows folder for {dataset_name}: {dataset_random_rows}")
			continue

		for pct in ["30", "60", "90"]:
			for health_status in ["healthy", "unhealthy"]:
				folder = os.path.join(dataset_random_rows, pct, health_status)
				if not os.path.isdir(folder):
					print(f"Missing folder: {folder}")
					continue

				expected_health = 1 if health_status == "healthy" else 0

				for fname in sorted(os.listdir(folder)):
					if not fname.endswith(".tsv"):
						continue

					path = os.path.join(folder, fname)
					df = pd.read_csv(path, sep='\t', dtype=str, low_memory=False)

					if "Sample" in df.columns:
						sample_ids = df["Sample"].astype(str)
					else:
						sample_ids = df.iloc[:, 0].astype(str)

					unique_samples = sample_ids.unique()
					missing_samples = [s for s in unique_samples if s not in meta_sample_set]
					wrong_group_samples = [
						s for s in unique_samples
						if s in meta_sample_set and health_map.get(s) != expected_health
					]

					results.append({
						"dataset": dataset_name,
						"percentage": pct,
						"group": health_status,
						"file": fname,
						"n_samples": len(unique_samples),
						"n_missing": len(missing_samples),
						"missing_samples": missing_samples,
						"n_wrong_group": len(wrong_group_samples),
						"wrong_group_samples": wrong_group_samples,
					})

					if verbose and missing_samples:
						print(
							f"{dataset_name} {pct}% {health_status} {fname}: "
							f"{len(missing_samples)} missing sample IDs"
						)
					if verbose and wrong_group_samples:
						print(
							f"{dataset_name} {pct}% {health_status} {fname}: "
							f"{len(wrong_group_samples)} samples in the wrong health group"
						)

	return pd.DataFrame(results)

results_df = validate_random_rows_samples("./", metadata_paths)
results_df.head()
```
