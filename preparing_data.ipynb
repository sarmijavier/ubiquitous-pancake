```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
# GETTING ALL THE MTB FILES AND REMOVING COLUMNS WITH ALL 0s

# 1. Define la ruta de la carpeta principal
ruta_base = Path('./processed_data')

# 2. Busca todos los archivos 'mtb.tsv' en las subcarpetas
# rglob busca de forma recursiva (en todas las carpetas hijo)
archivos_tsv = ruta_base.rglob('mtb.tsv')

# 3. Leer y combinar (opcional)
lista_dataframes = []

dfs_mtb = {}

for ruta in archivos_tsv:
    print(f"Leyendo: {ruta}")
    name = ruta.parent.name
    # Leemos con sep='\t' porque es un TSV (Tab Separated Values)
    df = pd.read_csv(ruta, sep='\t')
    print(f"Dimensiones antes de limpiar: {df.shape}")
    df = df.loc[:, (df != 0).any(axis=0)] # remove columns where all values are 0
    print(f"Dimensiones después de limpiar: {df.shape}")
    dfs_mtb[name] = df

```

# MAP MTB.MAP WITH MTB DATA

```python
# Remove NaN columns from HMDB, but keep KEGG if HMDB is NaN
dfs_mtb_renamed_clean = {}

#for dataset in interested_datasets:
for dataset in all_datasets:
	df = dfs_mtb[dataset].copy()
	map_df = dfs_mtp_map[dataset]
	
	# Create a mapping of old column names to new ones
	# If HMDB is NaN but KEGG exists, use KEGG; otherwise keep HMDB
	rename_dict = {}
	for idx, row in map_df.iterrows():
		old_name = row['Compound']
		hmdb = row['HMDB']
		kegg = row['KEGG']

		rename_dict[old_name] = hmdb
		if pd.isna(hmdb) and not pd.isna(kegg):
			# HMDB is NaN but KEGG is not, use KEGG
			rename_dict[old_name] = kegg

	# Rename columns based on the mapping
	df_renamed = df.rename(columns=rename_dict)
	
	dfs_mtb_renamed_clean[dataset] = df_renamed
```

```python

from sklearn.preprocessing import StandardScaler

# 1. Load your dataset
df = df_meta_clean

# 2. Set 'Sample' as index (important for merging with abundance data later)
df.set_index('Sample', inplace=True)

# 3. Drop administrative columns
cols_to_drop = ['Dataset', 'Subject', 'Age.Units', 'DOI', 'Publication.Name', 'Source Name']
df_clean = df.drop(columns=cols_to_drop)

# 4. One-Hot Encoding (Creates the "ENV1, ENV2" binary style)
# This will turn 'Gender' into 'Gender_Male' and 'Study.Group' into 'Study.Group_Control'
df_encoded = pd.get_dummies(df_clean, columns=['Study.Group', 'Gender'], drop_first=True)

# 5. Handle Missing Values (NAs)
# Machine learning models can't handle NAs. Impute with median for biological markers.
bio_markers = ['Age', 'BMI', 'Creatinine', 'Urea', 'eGFR', 'Dialysis.Frequency']
#df_encoded[bio_markers] = df_encoded[bio_markers].fillna(df_encoded[bio_markers].median())

# 6. Standardize (The "ENV6" floating-point style)
scaler = StandardScaler()
df_encoded[bio_markers] = scaler.fit_transform(df_encoded[bio_markers])

# This is now your "oneHotTest" formatted table
print(df_encoded.head())
```
