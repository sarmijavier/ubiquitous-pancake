```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
# GETTING ALL THE MTB FILES AND REMOVING COLUMNS WITH ALL 0s

# 1. Define la ruta de la carpeta principal
ruta_base = Path('./processed_data')

# 2. Busca todos los archivos 'mtb.tsv' en las subcarpetas
# rglob busca de forma recursiva (en todas las carpetas hijo)
archivos_tsv = ruta_base.rglob('mtb.tsv')

# 3. Leer y combinar (opcional)
lista_dataframes = []

dfs_mtb = {}

for ruta in archivos_tsv:
    print(f"Leyendo: {ruta}")
    name = ruta.parent.name
    # Leemos con sep='\t' porque es un TSV (Tab Separated Values)
    df = pd.read_csv(ruta, sep='\t')
    print(f"Dimensiones antes de limpiar: {df.shape}")
    df = df.loc[:, (df != 0).any(axis=0)] # remove columns where all values are 0
    print(f"Dimensiones después de limpiar: {df.shape}")
    dfs_mtb[name] = df

```

```python
# 1. Define la ruta de la carpeta principal
ruta_base = Path('./processed_data')

# 2. Busca todos los archivos 'genera.tsv' en las subcarpetas
# rglob busca de forma recursiva (en todas las carpetas hijo)
archivos_tsv = ruta_base.rglob('metadata.tsv')

dfs_metadata = {}

for ruta in archivos_tsv:
    print(f"Leyendo: {ruta}")
    name = ruta.parent.name  # Extrae el nombre de la carpeta padre
    # Leemos con sep='\t' porque es un TSV (Tab Separated Values)
    df = pd.read_csv(ruta, sep='\t')
    print(f"Dimensiones antes de limpiar: {df.shape}")
    df = df.loc[:, (df != 0).any(axis=0)] # remove columns where all values are 0
    print(f"Dimensiones después de limpiar: {df.shape}")
    dfs_metadata[name] = df

```

```python
# DICTIONARY TO GET THE COMPOUND NAME GIVEN THE HMDB ID
# {COMPOUND_NAME: HMDB_ID} }
# Then replace the column names with the HMDB id based on the dictionary

dfs_mtb_renamed = {}

interested_datasets = [
    'WANG_ESRD_2020', 
    'YACHIDA_CRC_2019', 
    'FRANZOSA_IBD_2019', 
    'ERAWIJANTARI_GASTRIC_CANCER_2020', 
    'iHMP_IBDMDB_2019', 
    'MARS_IBS_2020'
]

for dataset in interested_datasets:
    df = dfs_mtp_map[dataset]
    compound_hmdb_dict = dict(zip(df['Compound'], df['HMDB']))
    dfs_mtb_renamed[dataset] = dfs_mtb[dataset].rename(columns=compound_hmdb_dict)
```

```python
# Remove NaN columns from HMDB, but keep KEGG if HMDB is NaN
dfs_mtb_renamed_clean = {}

for dataset in interested_datasets:
	df = dfs_mtb[dataset].copy()
	map_df = dfs_mtp_map[dataset]
	
	# Create a mapping of old column names to new ones
	# If HMDB is NaN but KEGG exists, use KEGG; otherwise keep HMDB
	rename_dict = {}
	for idx, row in map_df.iterrows():
		old_name = row['Compound']
		hmdb = row['HMDB']
		kegg = row['KEGG']

		rename_dict[old_name] = hmdb
		if pd.isna(hmdb) and not pd.isna(kegg):
			# HMDB is NaN but KEGG is not, use KEGG
			rename_dict[old_name] = kegg

	# Rename columns based on the mapping
	df_renamed = df.rename(columns=rename_dict)
	
	dfs_mtb_renamed_clean[dataset] = df_renamed
```

```python
from itertools import combinations

# Get HMDB columns from each dataset mapping and find all possible combinations

hmdb_columns = {}

for dataset_name in interested_datasets:
	if dataset_name in dfs_mtp_map:
		map_df = dfs_mtp_map[dataset_name]
		if 'HMDB' in map_df.columns:
			hmdb_cols = set(map_df['HMDB'].dropna().unique())
			hmdb_columns[dataset_name] = hmdb_cols

# Find common HMDB for all possible combinations (including pairs)
all_combinations = []

for r in range(2, len(interested_datasets) + 1):
	for combo in combinations(interested_datasets, r):
		all_combinations.append(combo)

common_hmdb_all = {}

for combo in all_combinations:
	common = hmdb_columns[combo[0]].copy()
	
	for dataset in combo[1:]:
		common = common.intersection(hmdb_columns[dataset])
	
	if len(common) > 0:
		combo_name = " <-> ".join(combo)
		num_datasets = len(combo)
		common_hmdb_all[combo_name] = (len(common), num_datasets)

# Sort by number of common metabolites (descending), then by number of datasets (descending)
sorted_all = sorted(common_hmdb_all.items(), key=lambda x: (x[1][0], x[1][1]), reverse=True)

print("Common HMDB metabolites across all dataset combinations:\n")
for combo, (count, num_datasets) in sorted_all:
	print(f"[{num_datasets} datasets] {combo}: {count} common metabolites")
```

With these new data, I'll change the columns name and try to get HDBM or KEGG common columns

```python
import pandas as pd

# Remove NaN columns from HMDB, but keep KEGG if HMDB is NaN
dfs_mtb_renamed_clean = {}

for dataset in dataset_most_common_metabolites:
    df = dfs_mtb[dataset].copy()
    map_df = dfs_mtp_map[dataset]
    
    # Create a mapping of old column names to new ones
    rename_dict = {}
    for idx, row in map_df.iterrows():
        old_name = row['Compound']
        hmdb = row['HMDB']
        kegg = row['KEGG']

        # Determine the new name: HMDB preferred, then KEGG
        if pd.notna(hmdb):
            rename_dict[old_name] = hmdb
        elif pd.notna(kegg):
            rename_dict[old_name] = kegg
        else:
            # If both are NaN, we map it to None/NaN so we can drop it later
            rename_dict[old_name] = pd.NA

    # 1. Rename columns based on the mapping
    df_renamed = df.rename(columns=rename_dict)

    # 2. Delete any column where the name is NaN/Null
    # This keeps only columns where the column name itself is not null
    df_renamed = df_renamed.loc[:, df_renamed.columns.notna()]

    dfs_mtb_renamed_clean[dataset] = df_renamed
```

```python
# Create a new dataframe with the specified columns
combined_data = []

for dataset_name in dataset_most_common_metabolites:
	if dataset_name != 'YACHIDA_CRC_2019':
		continue

	mtb_df = dfs_mtb_renamed_clean[dataset_name]
	genera_df = dfs_genera_with_most_metabolites[dataset_name]
	metadata_df = dfs_metadata_with_most_metabolites[dataset_name]
	
	# Get common metabolites for this dataset
	metabolite_cols = [col for col in mtb_df.columns if col != 'Sample']
	genera_cols = [col for col in genera_df.columns if col != 'Sample']
	
	sample_count = 0
	
	for idx, mtb_row in mtb_df.iterrows():
		if sample_count >= 100:
			break
			
		sample_id = mtb_row['Sample']
		
		# Get metadata for this sample
		metadata_row = metadata_df[metadata_df['Sample'] == sample_id]
		if metadata_row.empty:
			continue
		
		status = metadata_row.iloc[0]['Subject']
		
		# Get genera data for this sample
		genera_row = genera_df[genera_df['Sample'] == sample_id]
		if genera_row.empty:
			continue
		
		# Iterate through bacteria and metabolites
		for bacteria in genera_cols:
			abundancia_genera = genera_row.iloc[0][bacteria]
			
			for metabolite in metabolite_cols:
				abundancia_mtb = mtb_row[metabolite]
				
				combined_data.append({
					'Study': dataset_name,
					'ID': sample_id,
					'Status': status,
					'Bacteria': bacteria,
					'abundancia_genera': abundancia_genera,
					'Metabolite': metabolite,
					'abundancia_mtb': abundancia_mtb
				})
				
#				if len(combined_data) >= 100:
#					break
			
#			if len(combined_data) >= 100:
#				break
		
		print(f"Processed sample {sample_id} from dataset {dataset_name}")
		sample_count += 1

combined_df = pd.DataFrame(combined_data)
print(f"Combined dataframe shape: {combined_df.shape}")
print(combined_df.head(10))
```
