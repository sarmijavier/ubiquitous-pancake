```python
def filter_genera(df):
    n_samples = len(df)
    presence = (df > 0).sum(axis=0) / n_samples  # fraction of samples where genus is present
    mean_abundance = df.mean(axis=0)
    keep = (presence >= 0.30) & (mean_abundance > 1e-4)
    return df.loc[:, keep]
```

```python
def create_random_subsets_rows(df, dataset_name):
    total_rows = df.shape[0]

    percentages = [0.30, 0.60, 0.90]
    num_iterations = 100

    for p in percentages:
        # Create directory
        dir_path_healthy = f"{dataset_name}/random_rows/{int(p*100)}/healthy"
        dir_path_unhealthy = f"{dataset_name}/random_rows/{int(p*100)}/unhealthy"

        df_main_meta = pd.read_csv(f"../processed_data/{dataset_name}/metadata.tsv", sep='\t')
        df_main_meta = df_main_meta.fillna(0)
        os.makedirs(dir_path_healthy, exist_ok=True)
        os.makedirs(dir_path_unhealthy, exist_ok=True)

        if dataset_name == "iHMP_IBDMDB_2019":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()
            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'nonIBD').astype(int)
            # 3. Finalize the meta_sub selection
            
        elif dataset_name == "WANG_ESRD_2020":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

            # This is the Python equivalent of your Julia list comprehension
            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Control').astype(int)
        
        elif dataset_name == "YACHIDA_CRC_2019":
            meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

            # This is the Python equivalent of your Julia list comprehension
            meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Healthy').astype(int)

        meta_sub = meta_sub[['Sample', 'is_healthy']]

        for i in range(1, num_iterations + 1):
            # 1. Number of rows to select
            num_to_select = int(round(total_rows * p))

            # 2. Sample random row indices (no replacement)
            selected_indices = np.random.choice(
                df.index,
                size=num_to_select,
                replace=False
            )

            # 3. Subset dataframe
            sub_df = df.loc[selected_indices, :]

            combined_df = pd.merge(sub_df, meta_sub, on='Sample', how='left')

            # get healthy and unhealthy subsets
            df_healthy = combined_df[combined_df['is_healthy'] == 1].copy()
            df_unhealthy = combined_df[combined_df['is_healthy'] == 0].copy()

            # Drop the helper column from both
            df_healthy = df_healthy.drop(columns=['is_healthy'])
            df_unhealthy = df_unhealthy.drop(columns=['is_healthy'])

            #4. Save file
            filename_healthy = f"sample_{int(p*100)}_run_{i}.tsv"
            filename_unhealthy = f"sample_{int(p*100)}_run_{i}.tsv"
            path_healthy = os.path.join(dir_path_healthy, filename_healthy)
            path_unhealthy = os.path.join(dir_path_unhealthy, filename_unhealthy)
            
            df_healthy.set_index('Sample', inplace=True)
            df_unhealthy.set_index('Sample', inplace=True)

            df_healthy = filter_genera(df_healthy)
            df_unhealthy = filter_genera(df_unhealthy)

            df_healthy.to_csv(path_healthy, sep="\t", index=True)
            df_unhealthy.to_csv(path_unhealthy, sep="\t", index=True)

            # Print only first iteration
            if i == 1:
                print(
                    f"{dataset_name}: "
                    f"Created {int(p*100)}% Row Subset (Run {i}) "
                    f"with {sub_df.shape[0]} samples and {sub_df.shape[1]} columns."
                )
```

```python
# cols In this case means samples. 
def create_random_subsets_cols(df, dataset_name):
	# 1. ID column (first column)
	id_col_name = df.columns[0]

	# 2. All other columns
	all_data_cols = df.columns[1:]

	percentages = [0.30, 0.60, 0.90]
	num_iterations = 100

	for p in percentages:
		# Create directory
		dir_path_healthy = f"{dataset_name}/random_columns/{int(p*100)}/healthy"
		dir_path_unhealthy = f"{dataset_name}/random_columns/{int(p*100)}/unhealthy"

		df_main_meta = pd.read_csv(f"../processed_data/{dataset_name}/metadata.tsv", sep='\t')
		os.makedirs(dir_path_healthy, exist_ok=True)
		os.makedirs(dir_path_unhealthy, exist_ok=True)

		if dataset_name == "iHMP_IBDMDB_2019":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()
			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'nonIBD').astype(int)
			# 3. Finalize the meta_sub selection
			
		elif dataset_name == "WANG_ESRD_2020":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

			# This is the Python equivalent of your Julia list comprehension
			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Control').astype(int)

		elif dataset_name == "YACHIDA_CRC_2019":
			meta_sub = df_main_meta[['Sample', 'Study.Group']].copy()

			# This is the Python equivalent of your Julia list comprehension
			meta_sub['is_healthy'] = (meta_sub['Study.Group'] == 'Healthy').astype(int)

		meta_sub = meta_sub[['Sample', 'is_healthy']]

		for i in range(1, num_iterations + 1):
			# 3. Number of columns to select
			num_to_select = int(round(len(all_data_cols) * p))

			# 4. Sample column names (no replacement)
			selected_data_cols = np.random.choice(
				all_data_cols,
				size=num_to_select,
				replace=False
			)

			# 5. Combine ID column + sampled columns
			final_selection = [id_col_name] + list(selected_data_cols)

			# 6. Subset dataframe
			sub_df = df.loc[:, final_selection]

			combined_df = pd.merge(sub_df, meta_sub, on='Sample', how='left')

			# get healthy and unhealthy subsets
			df_healthy = combined_df[combined_df['is_healthy'] == 1].copy()
			df_unhealthy = combined_df[combined_df['is_healthy'] == 0].copy()

			# Drop the helper column from both
			df_healthy = df_healthy.drop(columns=['is_healthy'])
			df_unhealthy = df_unhealthy.drop(columns=['is_healthy'])

			# 4. Save file
			filename_healthy = f"sample_{int(p*100)}_run_{i}.tsv"
			filename_unhealthy = f"sample_{int(p*100)}_run_{i}.tsv"
			path_healthy = os.path.join(dir_path_healthy, filename_healthy)
			path_unhealthy = os.path.join(dir_path_unhealthy, filename_unhealthy)

			df_healthy.set_index('Sample', inplace=True)
			df_unhealthy.set_index('Sample', inplace=True)
			df_healthy.to_csv(path_healthy, sep="\t", index=True)
			df_unhealthy.to_csv(path_unhealthy, sep="\t", index=True)

			# Print progress
			if i == 1:
				print(
					f"Created {int(p*100)}% subset (Run {i}) "
					f"with {sub_df.shape[1]} columns."
				)
```

```python
create_random_subsets_cols(df_iHMP_filtered, "iHMP_IBDMDB_2019")
create_random_subsets_cols(df_WANG_filtered, "WANG_ESRD_2020")
create_random_subsets_cols(df_YACHIDA_filtered, "YACHIDA_CRC_2019")
```

```python
import pandas as pd

INPUT_FILE = "../processed_data/YACHIDA_CRC_2019/genera.tsv"
OUTPUT_FILE ="../processed_data/YACHIDA_CRC_2019/genera_filtered.tsv"

PREVALENCE_THRESHOLD = 0.30   # present in at least 30% of samples
ABUNDANCE_THRESHOLD = 1e-4    # mean relative abundance > 10^-4
 
df = pd.read_csv(INPUT_FILE, sep="\t", index_col=0)

n_samples, n_genera = df.shape
print(f"Original: {n_samples} samples x {n_genera} genera")

# Prevalence filter: fraction of samples where genus > 0
prevalence = (df > 0).mean(axis=0)
prevalence_mask = prevalence >= PREVALENCE_THRESHOLD

# Abundance filter: mean relative abundance across all samples
mean_abundance = df.mean(axis=0)
abundance_mask = mean_abundance > ABUNDANCE_THRESHOLD

combined_mask = prevalence_mask & abundance_mask
df_filtered = df.loc[:, combined_mask]

n_kept = combined_mask.sum()
print(f"After filtering: {n_samples} samples x {n_kept} genera")
print(f"  - Passed prevalence (>={PREVALENCE_THRESHOLD*100:.0f}%): {prevalence_mask.sum()}")
print(f"  - Passed abundance (>{ABUNDANCE_THRESHOLD}): {abundance_mask.sum()}")
print(f"  - Passed both: {n_kept}")

df_filtered.to_csv(OUTPUT_FILE, sep="\t")
print(f"\nSaved to: {OUTPUT_FILE}")
```

```python
import pandas as pd
import numpy as np

PREVALENCE_THRESHOLD = 0.50   # present in at least 50% of samples (raised from 30%)
ABUNDANCE_THRESHOLD = 1e-4    # mean relative abundance > 10^-4
MIN_SAMPLES = 10              # minimum non-zero samples required per genus

# --- Load data ---
genera = pd.read_csv(
    "/Volumes/MZ BACKUP/data/processed_data/iHMP_IBDMDB_2019/genera.tsv", sep="\t", index_col="Sample"
)
metadata = pd.read_csv(
    "/Volumes/MZ BACKUP/data/processed_data/iHMP_IBDMDB_2019/metadata.tsv", sep="\t"
)

# --- Identify healthy samples ---
df_healthy_samples = metadata[metadata["Study.Group"] == "nonIBD"][["Sample", "Study.Group"]]
print(f"Healthy samples (total): {len(df_healthy_samples)}")
print(f"Unhealthy samples (total): {len(metadata) - len(df_healthy_samples)}")

# --- Merge genera with healthy/unhealthy label ---
metadata_slim = metadata[["Sample", "Study.Group"]].set_index("Sample")
full_df = genera.join(metadata_slim, how="inner")
print(f"Total samples after merge: {len(full_df)}")

# --- Take 30% random sample ---
random_30 = full_df.sample(frac=0.30, random_state=42)
print(f"\n30% random subset: {len(random_30)} samples")

# --- Split into healthy and unhealthy ---
random_healthy = random_30[random_30["Study.Group"] == "nonIBD"].drop(columns=["Study.Group"])
random_unhealthy = random_30[random_30["Study.Group"] != "nonIBD"].drop(columns=["Study.Group"])
print(f"  Healthy: {len(random_healthy)} samples")
print(f"  Unhealthy: {len(random_unhealthy)} samples")

# --- Filter genera per group ---
def get_passing_genera(df, label):
    n_samples = len(df)
    presence_frac = (df > 0).sum(axis=0) / n_samples
    mean_abundance = df.mean(axis=0)
    nonzero_count = (df > 0).sum(axis=0)

    keep = (
        (presence_frac >= PREVALENCE_THRESHOLD) &
        (mean_abundance > ABUNDANCE_THRESHOLD) &
        (nonzero_count >= MIN_SAMPLES)
    )
    passing = keep[keep].index
    print(f"\n  [{label}] n={n_samples} samples")
    print(f"    Passed prevalence (>={PREVALENCE_THRESHOLD*100:.0f}%): {(presence_frac >= PREVALENCE_THRESHOLD).sum()}")
    print(f"    Passed abundance (>{ABUNDANCE_THRESHOLD}): {(mean_abundance > ABUNDANCE_THRESHOLD).sum()}")
    print(f"    Passed min samples (>={MIN_SAMPLES}): {(nonzero_count >= MIN_SAMPLES).sum()}")
    print(f"    Passed all filters: {len(passing)}")
    return passing

print("\n--- Filtering genera per group ---")
healthy_genera = get_passing_genera(random_healthy, "healthy")
unhealthy_genera = get_passing_genera(random_unhealthy, "unhealthy")

# --- Keep only genera that pass in BOTH groups ---
shared_genera = healthy_genera.intersection(unhealthy_genera)
print(f"\nGenera passing in both groups: {len(shared_genera)}")

filtered_healthy = random_healthy[shared_genera]
filtered_unhealthy = random_unhealthy[shared_genera]

# --- Transpose: samples as columns, genera as rows ---
out_healthy = filtered_healthy.T
out_unhealthy = filtered_unhealthy.T

print(f"\nFinal output shapes (genera x samples):")
print(f"  Healthy:   {out_healthy.shape}")
print(f"  Unhealthy: {out_unhealthy.shape}")

# --- Save ---
out_healthy.to_csv("df_healthy_abundant_genera.tsv", sep="\t")
out_unhealthy.to_csv("df_unhealthy_abundant_genera.tsv", sep="\t")
print("\nSaved: df_healthy_abundant_genera.tsv")
print("Saved: df_unhealthy_abundant_genera.tsv")
```
